```python
from jiwer import wer, cer

# File paths
gt_path = "/content/wer_test_actual.txt"       # Ground truth file
pred_path = "/content/wer_test_predicted.txt"       # OCR output file

# Read files
with open(gt_path, "r", encoding="utf-8") as f:
    gt_text = f.read().strip()

with open(pred_path, "r", encoding="utf-8") as f:
    pred_text = f.read().strip()

# Compute metrics
word_error_rate = wer(gt_text, pred_text)
char_error_rate = cer(gt_text, pred_text)

print(" OCR Evaluation Results")
print("--------------------------------------")
print(f" Character-level Accuracy(CER):   {(1 - char_error_rate) * 100:.2f}%")
print(f" Word-level Accuracy(WER):        {(1 - word_error_rate) * 100:.2f}%")
```
